```python
import numpy as np
import pandas as pd

from scipy import stats
import json
```

```python
def read_beboc_results(path):
    df = pd.DataFrame([])
    for fold in range(5):
        with open(path+str(fold)+'.json', 'r') as f:
            fold_results = json.load(f)
        for tag in fold_results.keys():
            if tag not in df.index:
                df['label'] = list(fold_results.keys())
                df.set_index('label', drop=True, inplace=True)
            df.loc[tag, fold] = fold_results[tag]['f1']
    df['mean'] = df.mean(axis=1)
    df['std'] = df.std(axis=1)
    return df
```

```python
read_beboc_results('New_results/beboc_5folds_half_train_first/BEboC-5fold_tag_results_fold')
```

```python
def read_bebic_results(path):
    df = pd.DataFrame([])
    for fold in range(5):
        with open(path+str(fold)+'.json', 'r') as f:
            fold_results = json.load(f)
        best_mean_f1 = 0
        best_epoch = 0
        for i, f1s in enumerate(zip(fold_results['ORG']['f1'], fold_results['LOC']['f1'], fold_results['PER']['f1'])):
            mean_f1 = np.mean(f1s)
            if mean_f1 > best_mean_f1:
                best_mean_f1 = mean_f1
                best_epoch = i
        for tag in fold_results.keys():
            if tag not in df.index:
                df['label'] = list(fold_results.keys())
                df.set_index('label', drop=True, inplace=True)
            df.loc[tag, fold] = fold_results[tag]['f1'][best_epoch]
    df['mean'] = df.mean(axis=1)
    df['std'] = df.std(axis=1)
    return df
```

```python
read_bebic_results('New_results/bebic_5folds_half_train_first/BEbiC-5fold_head_results_fold')
```
